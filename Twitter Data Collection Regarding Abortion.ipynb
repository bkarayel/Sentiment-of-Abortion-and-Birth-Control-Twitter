### Twitter Data Collection Regarding Abortion

```python
import os
import pandas as pd
```

```python
tweet_count = 500000
text_query = "kürtaj"
since_date = "2012-05-12"
until_date = "2022-05-12"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> berrakürtajson.json'.format(tweet_count, since_date, text_query, until_date))

# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_kürtaj = pd.read_json('berrakürtajson.json', lines=True)

# Displays first 5 entries from dataframe
tweets_kürtaj
```

```python
abortion_tweets=tweets_kürtaj[['date','rawContent','id','hashtags']]
```

```python
abortion_tweets
```

```python
abortion_tweets.rename(columns = {'date':'Date','rawContent':'Tweet Text','id':'Tweet ID','hashtags':'Hashtags'}, inplace = True)
```

```python
abortion_tweets.to_csv('abortiontweets.csv', sep=',', index=False)
```

```python
abortion_tweets_datatoexcel = pd.ExcelWriter('abortiontweets.xlsx')
```

```python
abortion_tweets.to_excel(abortion_tweets_datatoexcel)
```
